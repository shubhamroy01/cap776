```python
# mathmatical functions: Arithmetic Functions
import numpy as np
a = np.array([10,20,30])
b= np.array([2,3,4])
print("addition of a and b :",np.add(a,b))
print("subtraction of a and b :",np.subtract(a,b))
print("multiplication of a and b :",np.multiply(a,b))
print("division of a and b :",np.divide(a,b))
print("remainders when a/b :",np.remainder(a,b))
print("a raised to b is:",np.power(a,b))
print("mod of a and b :",np.mod(a,b))
```

```python
# mathmatical functions:Trigonometric Functions
import numpy as np
from matplotlib import pyplot as plt 
print(np.pi)
a = np.array([0,np.pi/6,np.pi/4,np.pi/3,np.pi/2]) 
degreeA = np.degrees(a)
print(degreeA)
sin_degree = np.sin(a)
print("sin values",sin_degree)
cos_degree = np.cos(a)
print("cos values",cos_degree)
tan_degree = np.tan(a)
print("tan values",tan_degree)
print(np.arcsin(sin_degree))
# Compute the x and y coordinates for points on a sine curve 
x = np.arange(0, 3 * np.pi, 0.1) 
y = np.sin(x) 
plt.title("sine wave form") 

# Plot the points using matplotlib 
plt.plot(x, y) 
plt.show() 
```

```python
# mathmatical functions:Logarithmic and Exponential Functions
import numpy as np
a = np.array([1,2,3,4,5])
a_log = np.log(a)
a_exp = np.exp(a)
print("log of input array a is:",a_log)
print("exponent of input array a is:",a_exp)
```

```python
# mathmatical functions:Rounding Functions
import numpy as np
a = np.array([1.23,4.165,3.8245])
rounded_a = np.round_(a,2)
print(rounded_a)
floor_a = np.floor(a)
print(floor_a)
```

```python
import numpy as np
from matplotlib import pyplot as plt 
x = np.arange(0,4*np.pi-1,0.1)   # start,stop,step
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 4pi')  # string must be enclosed with quotes '  '
plt.ylabel('sin(x) and cos(x)')
plt.title('Plot of sin and cos from 0 to 4pi')
plt.legend(['sin(x)', 'cos(x)'])      # legend entries as seperate strings in a list
plt.show()
```

```python
import numpy as np  
  
a = np.array([[10,2,3],[4,5,6],[7,8,9]])  
  
print("Sorting along the columns:")  
print(np.sort(a))  
  
print("Sorting along the rows:")  
print(np.sort(a, 0))  

# Order parameter in sort function 
dt = np.dtype([('name', np.str_,16),('age', int)]) 
arr = np.array([("raju",21),("anil",25),("ravi", 17), ("amar",27)], dtype = dt)
print("Sorting data ordered by name")  
print(np.sort(arr,order = 'name'))  
```
