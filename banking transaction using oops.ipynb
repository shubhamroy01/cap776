```python
balance=1000
class transaction:
    
    def checkbalance(self):
        return balance
    def deposit(self,damt):
        global balance
        balance=balance+damt
        return balance
    def withdrawal(self,wamt):
        global balance
        if wamt<=balance-1000:
            balance=balance-wamt
            return balance
        else:
            return 0

while(True):
    print("Select your choice\n1.Balance\n2.Deposit\n3.withdrawal\n4.Exit\n")
    ch=int(input())
    t1=transaction()
    if ch==1:
        print("Balance:{}".format(t1.checkbalance()))
    if ch==2:
        amt=int(input("Enter amount\n"))
        print("Balance:{}".format(t1.deposit(amt)))
    if ch==3:
        amt=int(input("Enter amount\n"))
        waamt=t1.withdrawal(amt)
        if(waamt==0):
            print("Insufficient Balance")
        else:
            print("Balance:{}".format(waamt))
           
    if ch==4:
        print("Thanks visit again")
        break
    
```
